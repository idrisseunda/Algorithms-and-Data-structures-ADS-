```python
def strg2(s):
    n = len(s)
    f = []
    l = 0
    u = 0
    
    # First pass: find invalid ')'
    for i in range(n):
        if s[i] == "(":
            l = l + 1
        elif s[i] == ")":
            u = u + 1
            if u > l:
                f.append(i)
                u = u - 1  # Don't count this invalid ')'
    
    # Remove invalid ')' first
    s = sm(s, f)
    
    # Second pass: find invalid '('
    n = len(s)
    f = []
    v = 0
    k = 0
    
    for i in range(n):
        if s[n - 1 - i] == "(":
            v = v + 1
            if v > k:
                f.append(n - i - 1)
                v = v - 1  # Don't count this invalid '('
        elif s[n - 1 - i] == ")":
            k = k + 1
    
    # Remove invalid '(' and return
    s = sm(s, f)
    return s  # âœ“ Always returns
```
